```python
# loading all the module needed #
import seaborn as sms
import numpy as np
import matplotlib.pyplot as plt
```

```python
# input data you want to work with #
data = np.random.rand(10, 20)
```

```python
sms.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```
